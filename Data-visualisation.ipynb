```python
##Comparison between Decision Tree and Neural Network for Data Classification
```

```python
#Install sklearn pip install sklearn 
import numpy as np 
import matplotlib.pyplot as pt 
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.neural_network import MLPClassifier 

from sklearn.neighbors import NearestNeighbors 
# Read data 
data=pd.read_csv("train.csv").values 
print(data) 
print(data.shape) #42000 
#Split Data into train and test dataset 
xtrain=data[0:21000,1:] 
print(xtrain) 
train_label=data[0:21000,0] 
print(train_label) 
xtest=data[21000:,1:] 
test_label=data[21000:,0] 
#showing an input image 
d=xtest[8] 
d.shape=(28,28) 
pt.imshow(255-d,cmap='gray') 
# Initialize Classifier  
clf=DecisionTreeClassifier() 
#Training 
clf.fit(xtrain,train_label) 
#Testing 1 image 
 
print(clf.predict([xtest[8]])) 
pt.show() 
#Testing Overall dataset of test 
p=clf.predict(xtest) 
#Accuracy 
count=0 
for i in range(21000): 
     if p[i]==test_label[i]:count+=1 
print("Accuracy of Decision Tree = {}".format((count/21000)*100)) 

clf1 = MLPClassifier(activation='logistic',solver='sgd',hidden_layer_sizes=(28,10),random_state=1) 
clf1.fit(xtrain,train_label) 
q=clf1.predict(xtest) 
count=0 
for i in range(21000): 
    if q[i]==test_label[i]:count+=1  
print("Accuracy of Neural Network = {}".format(count/21000)) 
```
